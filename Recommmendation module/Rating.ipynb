```python
user_id=1
product_id=11

total_days=13
product_days=7

def ratings(total_days,product_days):
    x=product_days/total_days
    rating=0
    if x<0.2:
        rating=1
    elif 0.2<x<0.4:
        rating=2
    elif 0.4<x<0.6:
        rating=3
    elif 0.6<x<0.8:
        rating=4
    elif 0.8<x<=1:
        rating=5
    return rating
```

```python
rating=ratings(total_days,product_days)
train_df.loc([(train_df['user_id']==user_id) & (train_df['product_id']==product_id)],'rating')=rating
```
