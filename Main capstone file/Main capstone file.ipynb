```python
import cv2
import numpy as np
import os
import torch
from torchvision import transforms,models,datasets
import torchvision

torch.cuda.empty_cache()
```

```python
class_names=['Dhruv', 'Devansh', 'Urmila']
device=torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
face_cascade = cv2.CascadeClassifier("C:/Users/hp/Desktop/Capstone/haarcascade_frontalface_default.xml")
model = torchvision.models.resnet18(num_classes=len(class_names)).eval().to(device)
model.load_state_dict(torch.load('model_weights.pth'))
```

```python
def face_extractor(img):
    global cropped_face
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    if faces is ():
        return None

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face
```

```python
transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
video_capture = cv2.VideoCapture(0)
count=0
while True:
    _,frame = video_capture.read()
    '''Extract face , convert to grayscale and save it in out folders'''
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (244, 244))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        file_name_path = 'C:/Users/hp/Desktop/Pytorch/buffer/' + str(count) + '.jpg'

        cv2.imwrite(file_name_path, face)
        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow('Face Cropper', face)
    else:
        print("Face not found")
        pass
    if cv2.waitKey(1) == 13 or count == 100:
        break
video_capture.release()
cv2.destroyAllWindows()
```

```python
data_dir='C:/Users/hp/Desktop/Pytorch/buffer/'
buffer_data=datasets.ImageFolder(data_dir,transform=transform)
buffer_loader=torch.utils.data.DataLoader(buffer_data,batch_size=4)
```
